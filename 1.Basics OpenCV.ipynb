# Blank Image and drawing on it

```python

blank_image = np.zeros(shape=(512,512,3),dtype=np.int16)
print(blank_image.shape)
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(384,0),pt2=(150,150),color=(0,255,0),thickness=10)
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(200,200),pt2=(300,300),color =(255,255,255),thickness =-10)
plt.imshow(blank_image)
```

```python
cv2.circle(img = blank_image,center =(100,100),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_image)
```

```python
cv2.line(blank_image,pt1=(0,0),pt2=(300,300),color =(255,55,40),thickness=5)
plt.imshow(blank_image)
```

```python
font = cv2.FONT_ITALIC
cv2.putText(blank_image,text='Hello Sajid',org=(10,500),fontFace=font,
            fontScale=1,color=(255,255,255),
            thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_image)
```

```python
#Polygon
blank_image=np.zeros(shape=(512,512,3),dtype=np.int32)
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
pts = vertices.reshape((-1,1,2)) #convert for 3 channel

cv2.polylines(blank_image,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_image)
```
